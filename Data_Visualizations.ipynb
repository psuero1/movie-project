```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statistics
```

```python
gross_2018 = pd.read_csv('top_gross_2018', index_col = 0)
gross_2018
```

```python
gross_2017 = pd.read_csv('top_gross_2017', index_col = 0)
gross_2016 = pd.read_csv('top_gross_2016', index_col = 0)
gross_2015 = pd.read_csv('top_gross_2015', index_col = 0)
gross_2014 = pd.read_csv('top_gross_2014', index_col = 0)
```

```python
runtime_mean18 = statistics.mean(gross_2018['runtime_minutes'].tolist())
runtime_mean17 = statistics.mean(gross_2017['runtime_minutes'].tolist())
runtime_mean16 = statistics.mean(gross_2016['runtime_minutes'].tolist())
runtime_mean15 = statistics.mean(gross_2015['runtime_minutes'].tolist())
runtime_mean14 = statistics.mean(gross_2014['runtime_minutes'].tolist())
```

## Creating bar graph to show Run Times

```python
## Creating lists to plot runtimes subplots to compare years
runtime_2018 = gross_2018['runtime_minutes'].tolist()
runtime_2017 = gross_2017['runtime_minutes'].tolist()
runtime_2016 = gross_2016['runtime_minutes'].tolist()
runtime_2015 = gross_2015['runtime_minutes'].tolist()
runtime_2014 = gross_2014['runtime_minutes'].tolist()
titles_2018 = gross_2018['title'].tolist()
titles_2017 = gross_2017['title'].tolist()
titles_2016 = gross_2016['title'].tolist()
titles_2015 = gross_2015['title'].tolist()
titles_2014 = gross_2014['title'].tolist()
```

```python
x1 = titles_2018
x2 = titles_2017
x3 = titles_2016
x4 = titles_2015
x5 = titles_2014

plt.figure(figsize =(16, 8))
plt.subplot(2,3,1)
title18 = plt.bar(x1, runtime_2018)
plt.ylabel('Runtime in Minutes')
plt.xticks(x1, rotation = 'vertical')
plt.title('Runtimes of Top Grossing 2018 Movies')

plt.subplot(2,3,2)
title17 = plt.bar(x2, runtime_2017)
plt.ylabel('Runtime in Minutes')
plt.xticks(x2, rotation = 'vertical')
plt.title('Runtimes of Top Grossing 2017 Movies')

plt.subplot(2,3,3)
title16 = plt.bar(x3, runtime_2016)
plt.ylabel('Runtime in Minutes')
plt.xticks(x3, rotation = 'vertical')
plt.title('Runtimes of Top Grossing 2016 Movies')



plt.show()
```

```python
plt.figure(figsize =(20,12))
plt.subplot(2,3,1)
title15 = plt.bar(x4, runtime_2015)
plt.ylabel('Runtime in Minutes')
plt.xticks(x4, rotation = 'vertical')
plt.title('Runtimes of Top Grossing 2015 Movies')

plt.subplot(2,3,2)
title14 = plt.bar(x5, runtime_2014)
plt.ylabel('Runtime in Minutes')
plt.xticks(x5, rotation = 'vertical')
plt.title('Runtimes of Top Grossing 2014 Movies')
```

```python
plt.figure(figsize=(9,3))
x = ('2014', '2015', '2016', '2017', '2018')
y = (runtime_mean14, runtime_mean15, runtime_mean16, runtime_mean17, runtime_mean18)

plt.figure(figsize =(16, 8))
runtime = plt.bar(x, y )
plt.ylabel('Runtime in Minutes')
plt.xticks(x)
plt.title('Average Runtimes of Top Grossing Movies 2014-2018')

```

## Average run times havent changed much in the past 5 years

## Looking at ratings

```python
all_movies = pd.read_csv('final.csv', index_col = 0)
all_movies.head()
```

```python
top_rated18 = all_movies[(all_movies['release_date'] >= '2018')].sort_values(by ='averagerating', ascending = False)
top_rated18.head()
```

```python
top_rated18.tail()
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Ratings')
ax1.boxplot(top_rated18['averagerating'], vert= False, showfliers = False)

fig1, ax2 = plt.subplots()
ax2.set_title('Domestic Gross')
ax2.boxplot(top_rated18['domestic_gross'],vert = False, showfliers = False)

fig1, ax3 = plt.subplots()
ax3.set_title('Runtime')
ax3.boxplot(top_rated18['runtime_minutes'], vert =False, showfliers = False)
```
