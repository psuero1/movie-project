```python
import pandas as pd
```

## Looking at all the data sets

```python
tmbd_mov = pd.read_csv('zippedData/tmdb.movies.csv.gz', encoding='latin_1')
tmbd_mov.head()
```

```python
tmbd_mov.tail()
```

```python
tmbd_mov_titles = tmbd_mov[['title', 'release_date']]
tmbd_mov_titles.head()
```

```python
imdb_name = pd.read_csv('zippedData/imdb.name.basics.csv.gz', encoding='latin_1')
imdb_name.head()
```

```python
imdb_tbasics = pd.read_csv('zippedData/imdb.title.basics.csv.gz', encoding='latin_1')
imdb_tbasics.head()
```

```python
imdb_takas = pd.read_csv('zippedData/imdb.title.akas.csv.gz', encoding='latin_1')
imdb_takas.head()
```

```python
imdb_tratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz', encoding='latin_1')
imdb_tratings.head()
```

```python
rt_info = pd.read_csv('zippedData/rt.movie_info.tsv.gz', sep='\t')
rt_info.head()
```

```python
rt_reviews = pd.read_csv('zippedData/rt.reviews.tsv.gz', sep='\t', encoding='latin_1')
rt_reviews.head()
```

```python
imdb_tcrew = pd.read_csv('zippedData/imdb.title.crew.csv.gz', encoding='latin_1')
imdb_tcrew.head()
```

```python
imdb_tprincipals = pd.read_csv('zippedData/imdb.title.principals.csv.gz', encoding='latin_1')
imdb_tprincipals.head()
```

## Starting to merge multiple data frames

```python
imdb_titles = pd.merge(imdb_tbasics, imdb_tratings, on = 'tconst', how = 'outer')
imdb_titles.head()
```

```python
imdb_titles.tail()
```

```python
imdb_titles.isna().sum()
```

```python
imdb_movies = imdb_titles.rename(columns={'primary_title': 'title'})
imdb_movies.head()
```

## Merging df to add in release dates and titles from tmdb

```python
all_mov_titles = pd.merge(imdb_movies, tmbd_mov_titles, on = 'title', how = 'outer')
all_mov_titles.head()
```

```python
all_mov_titles.info()
```

```python
all_mov_titles.isna().sum()
```

## Testing what would happen if I drop all null values

```python
test_all = all_mov_titles.dropna()
test_all.tail()
```

```python
test_all.info()
```

```python
test_all.isna().sum()
```

```python
test_all.tail()
```

```python
## This looks good to me so far. I know there are duplicates. Going to look into merging the gross data in based on title
## I want to remove the start_year column, now that I have the release_date in there
## Will also remove year column in gross. 
## Then I can merge gross into my test_all using title and outer join 
```

```python
all_movies = test_all.drop(["start_year", "original_title"], axis=1)
all_movies.head()
```

## Looking to see how I can merge gross into my big df above

```python
bom_gross = pd.read_csv('zippedData/bom.movie_gross.csv.gz', encoding='latin_1')
bom_gross.head()
```

```python
bom_gross.isna().sum()
```

```python
bom_gross.tail()
```

## Removed year column from gross to prepare merge

```python
gross = bom_gross.drop(columns="year")
gross.head()
```

## Merging df's to have my final set for now

```python
all_mov_gross = pd.merge(all_movies, gross, on = 'title', how = 'outer')
all_mov_gross.head()
```

```python
all_mov_gross.tail()
```

```python
test_final = all_mov_gross.dropna()
test_final.head()
```

```python
test_final.info()
```

## Checking for duplicates

```python
test_final.duplicated('title')
```

```python
final = test_final.drop_duplicates(subset = 'title', keep = "first")
final.head()
```

```python
final.tail()
```

```python
final.info()
```
