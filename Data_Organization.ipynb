```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
all_movies = pd.read_csv('final.csv')
all_movies.head()
```

## Removing extra column from original df

```python
final_movies = all_movies.drop(['Unnamed: 0'], axis=1)
final_movies.head()
```

```python
final_movies.info()
```

## Finding newest movies by release date

```python
final_movies['release_date'] = pd.to_datetime(final_movies['release_date'])
by_release = final_movies.sort_values(by = "release_date", ascending = False)
by_release.head()
```

## Top Domestic Gross

```python
top_dgross = final_movies.sort_values(by='domestic_gross', ascending = False)
top_dgross.head(10)
```

## Top 10 domestic gross for the past 5 years

```python
top10_dgross_18 = final_movies[(final_movies['release_date'] >= "2018") & 
                             (final_movies['domestic_gross'] > 0)].sort_values(by = "domestic_gross", ascending = False).head(10)
top10_dgross_18
```

```python
top10_dgross_17 = final_movies[(final_movies['release_date'] <= "2018") & (final_movies['release_date'] >= "2017") &
                             (final_movies['domestic_gross'] > 0)].sort_values(by = "domestic_gross", ascending = False).head(10)
top10_dgross_17
```

```python
top10_dgross_16 = final_movies[(final_movies['release_date'] <= "2017") & (final_movies['release_date'] >= "2016") &
                               (final_movies['domestic_gross'] > 0)].sort_values(by = "domestic_gross", ascending = False).head(10)
top10_dgross_16
```

```python
top10_dgross_15 = final_movies[(final_movies['release_date'] <= "2016") & (final_movies['release_date'] >= "2015") &
                             (final_movies['domestic_gross'] > 0)].sort_values(by = "domestic_gross", ascending = False).head(10)
top10_dgross_15
```

```python
top10_dgross_14 = final_movies[(final_movies['release_date'] <= "2015") & (final_movies['release_date'] >= "2014") &
                             (final_movies['domestic_gross'] > 0)].sort_values(by = "domestic_gross", ascending = False).head(10)
top10_dgross_14
```
