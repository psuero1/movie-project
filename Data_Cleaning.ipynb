```python
imdb_name = pd.read_csv('zippedData/imdb.name.basics.csv.gz', encoding='latin_1')
imdb_name.head()
```

```python
imdb_tbasics = pd.read_csv('zippedData/imdb.title.basics.csv.gz', encoding='latin_1')
imdb_tbasics.head()
```

```python
imdb_takas = pd.read_csv('zippedData/imdb.title.akas.csv.gz', encoding='latin_1')
imdb_takas.head()
```

```python
imdb_tratings = pd.read_csv('zippedData/imdb.title.ratings.csv.gz', encoding='latin_1')
imdb_tratings.head()
```

```python
imdb_titles = pd.merge(imdb_tbasics, imdb_tratings, on = 'tconst', how = 'outer')
imdb_titles.head()
```

```python
imdb_titles.tail()
```

```python
imdb_titles.isna().sum()
```

```python
imdb_movies = imdb_titles.rename(columns={'primary_title': 'title'})
imdb_movies.head()
```

```python
imdb_mov_titles = pd.merge(imdb_movies, tmbd_mov, on = 'title', how = 'left')
imdb_mov_titles.head()
```

```python
imdb_mov_titles.tail()
```
